Get data from database (MySQL)

```python
class DATA:
    def last_season(ha):
        c = DB.cursor()
        c.execute(f"""select 
common_name,
matches_played_{ha},
goals_scored_{ha},
goals_conceded_{ha},
goal_difference_{ha},
goals_scored_per_match_{ha},
goals_conceded_per_match_{ha},
promotion,
season_id,
country
from teams_2122;""")
        DB1 = c.fetchall()
        DF = pd.DataFrame(DB1, columns=('Team','Played','F','A','diff','scored_avg','conceded_avg','promotion','season_id','country'))
        return DF
    def ELO():
        c = DB.cursor()
        c.execute("""select team_name, ELO
from elo_raking_2122;""")
        eloDB = c.fetchall()
        eloDF = pd.DataFrame(eloDB, columns=('Team','ELO'))
        return eloDF
    def matchesDF():
        c = DB.cursor()
        c.execute("""select 
	home_team_name,
    away_team_name,
    home_team_goal_count,
    away_team_goal_count,
    total_goal_count,
    season_id
from matches_2122;""")
        matchesDB = c.fetchall()
        matchesDF = pd.DataFrame(matchesDB, columns=('home_team','away_team','home_goals','away_goals','total_goals','season_id'))
        return matchesDF
```

Statistics calc

```python
class stat:
    def goals_last_season(team, ha, scored:bool):
        last_sea = DATA.last_season(f'{ha}')
        team_data = last_sea.loc[last_sea['Team']==team]
        htd = team_data
        if scored == True:
            if htd['promotion'].values[0]=='relegation':
                F1 = htd['F'].values[0]*1.6
                MP1 = htd['Played'].values[0]
                return team, MP1, F1
            if htd['promotion'].values[0]=='promotion':
                F1 = htd['F'].values[0]*0.6
                MP1 = htd['Played'].values[0]
                return team, MP1, F1
            elif htd['promotion'].values[0]=='stand':
                F1 = htd['F'].values[0]
                MP1 = htd['Played'].values[0]
                return team, MP1, F1
            else: return print('check promotion columns on DATA')
        if scored == False:
            if htd['promotion'].values[0]=='relegation':
                A1 = htd['A'].values[0]*0.6
                MP1 = htd['Played'].values[0]
                return team, MP1, A1
            if htd['promotion'].values[0]=='promotion':
                A1 = htd['A'].values[0]*1.6
                MP1 = htd['Played'].values[0]
                return team, MP1, A1
            elif htd['promotion'].values[0]=='stand':
                A1 = htd['A'].values[0]
                MP1 = htd['Played'].values[0]
                return team, MP1, A1
            else: return print('check promotion columns on DATA')
    def xG_stat(home_team, away_team):
        
        home_MP1 = stat.goals_last_season(home_team,'home',True)[1]
        home_goals1 = stat.goals_last_season(home_team,'home',True)[2]
        home_conc1 = stat.goals_last_season(home_team,'home',False)[2]
        away_MP1 = stat.goals_last_season(away_team,'away',True)[1]
        away_goals1 = stat.goals_last_season(away_team,'away',True)[2]
        away_conc1 = stat.goals_last_season(away_team,'away',False)[2]
        home_sc = home_goals1/home_MP1
        home_conc = home_conc1/home_MP1
        away_sc = away_goals1/away_MP1
        away_conc = away_conc1/away_MP1
        xG_home = (home_sc + away_conc)/2
        xG_away = (away_sc + home_conc)/2
        
        return xG_home, xG_away
    def elo_goals(home_team:str, away_team:str):
        s = DATA.last_season('home')
        season = s.loc[s['Team']==home_team]['season_id'].values[0]
        elo = DATA.ELO()
        eloh =  elo[elo['Team']==home_team]
        eloa =  elo[elo['Team']==away_team]
        elo_home = eloh['ELO'].values
        elo_away = eloa['ELO'].values

        dataframe = DATA.matchesDF()
        DF = dataframe.loc[dataframe['season_id']==season]
        total_avg = DF['home_goals'].mean()+DF['away_goals'].mean()
        elo1 = 1/(10**((elo_away[0]-elo_home[0])/400)+1)
        elo2 = 1/(10**((elo_home[0]-elo_away[0])/400)+1)
        xG_home = elo1 * total_avg
        xG_away = elo2 * total_avg
        return xG_home, xG_away, elo1, elo2
```

Factors Coeficients

```python
class new_coef:
    def zero_inf(home_team,away_team):
        s = DATA.last_season('home')
        season = s.loc[s['Team']==home_team]['season_id'].values[0]
        c3 = DB.cursor()
        c3.execute(f"""select 
	    home_team_name,
        away_team_name,
        home_team_goal_count,
        away_team_goal_count,
        total_goal_count
        from matches_2122
        where season_id = '{season}'""")
        mDB = c3.fetchall()
        matches_local = pd.DataFrame(mDB, columns=('home_team','away_team','home_goals','away_goals','total_goals'))
        zero = 0
        n=0
        for i in matches_local.values:
            a = matches_local.loc[n]
            n +=1
            if a['home_goals'] == 0 or a['away_goals'] == 0:
                zero += 1
            else:
                continue
        avg_home = matches_local['home_goals'].values.mean()
        avg_away = matches_local['away_goals'].values.mean()
        

        home_prb = []
        for i in np.arange(10):
            home_prb.append(poisson.pmf(k=i, mu=avg_home))
        

        away_prb = []
        for i in np.arange(10):
            away_prb.append(poisson.pmf(k=i, mu=avg_away)) 
      
        prod_table = np.array([(i*j) for i, j in product(home_prb, away_prb)])
        prod_table.shape = (10, 10)

        prob_df = pd.DataFrame(prod_table, index=range(10), columns=range(10))
        

        zero_result = [prob_df.iat[0,0],prob_df.iat[0,1],prob_df.iat[0,2],prob_df.iat[0,3],prob_df.iat[0,4],prob_df.iat[0,5],prob_df.iat[0,6],prob_df.iat[0,7],prob_df.iat[0,8],
        prob_df.iat[1,0],prob_df.iat[2,0],prob_df.iat[3,0],prob_df.iat[4,0],prob_df.iat[5,0],prob_df.iat[6,0],prob_df.iat[7,0],prob_df.iat[8,0],prob_df.iat[9,0]]
        
        poisson_0 = np.array(zero_result).sum()
        return zero/n - poisson_0
    def draw_coef(home_team:str, away_team:str):
        eloDF = DATA.ELO()
        t1 = eloDF.loc[eloDF['Team']==home_team]
        t2 = eloDF.loc[eloDF['Team']==away_team]
        elo1 = t1['ELO'].values
        elo2 = t2['ELO'].values
        if abs(elo1-elo2) <= 50:
            return 0.1
        if abs(elo1-elo2)>50 and abs(elo1-elo2) <= 150:
            return 0.08
        if abs(elo1-elo2)>150 and abs(elo1-elo2) <= 250:
            return 0.06
        if abs(elo1-elo2)>250 and abs(elo1-elo2) <= 350:
            return 0.03
        if abs(elo1-elo2)>350:
            return 0 
class adv:
    def home_adv(home_team):
        homeDF = DATA.last_season('home')
        ht = homeDF.loc[homeDF['Team']==home_team]
        if ht['promotion'].values[0]=='relegation':
                F_new = ht['F']*1.5/ht['Played']
                A_new = ht['A']*0.5/ht['Played']
                ha = F_new - A_new
        if ht['promotion'].values[0]=='promotion':
                F_new = ht['F']*0.5/ht['Played']
                A_new = ht['A']*1.5/ht['Played']
                ha = F_new - A_new
        if ht['promotion'].values[0]=='stand':
                F_new = ht['F']/ht['Played']
                A_new = ht['A']/ht['Played']
                ha = F_new - A_new
        return ha
    def away_adv(away_team):
        awayDF = DATA.last_season('away')
        ht = awayDF.loc[awayDF['Team']==away_team]
        if ht['promotion'].values[0]=='relegation':
                F_new = ht['F']*1.5/ht['Played']
                A_new = ht['A']*0.5/ht['Played']
                ha = F_new - A_new
        if ht['promotion'].values[0]=='promotion':
                F_new = ht['F']*0.5/ht['Played']
                A_new = ht['A']*1.5/ht['Played']
                ha = F_new - A_new
        if ht['promotion'].values[0]=='stand':
                F_new = ht['F']/ht['Played']
                A_new = ht['A']/ht['Played']
                ha = F_new - A_new
        return ha

    def home_advantage(home_team,away_team):
        a1 = adv.home_adv(home_team)
        a2 = adv.away_adv(away_team)
        a = (a1.values[0]-a2.values[0])
        return a/7

class new_expected_goals:
    def xG(home_team,away_team, values = True):
        stat_xG = stat.xG_stat(home_team, away_team)
        elo_xG = stat.elo_goals(home_team, away_team)
        home_xG = (stat_xG[0]+elo_xG[0])/2
        away_xG = (stat_xG[1]+elo_xG[1])/2
        if values == True:
            return float("{:.2f}".format(home_xG)), float("{:.2f}".format(away_xG))
        if values == False:
            return home_team + ' ' + str(home_xG), away_team + ' ' + str(away_xG)
```

```python
class new_prb:
    def prb_matrix(home_team:str, away_team:str):
        zero_inf = abs(new_coef.zero_inf(home_team,away_team))
        draw_coef= new_coef.draw_coef(home_team,away_team)
        a = new_expected_goals.xG(home_team,away_team)
        xG_home = a[0]+adv.home_advantage(home_team,away_team)
        xG_away = a[1]
        
        home_prb = []
        for i in np.arange(10):
            home_prb.append(poisson.pmf(k=i, mu=xG_home)) 
 
        away_prb = []
        for i in np.arange(10):
            away_prb.append(poisson.pmf(k=i, mu=xG_away))

        home_prb1 = [(zero_inf+(1-draw_coef)*(1-zero_inf)*home_prb[0]),(draw_coef+(1-draw_coef)*(1-zero_inf)*home_prb[1])]
        lst = home_prb[2::]
        for i in lst:
            home_prb1.append(i*(1-draw_coef)*(1-zero_inf))
        
        away_prb1 = [(zero_inf+(1-draw_coef)*(1-zero_inf)*away_prb[0]),(draw_coef+(1-draw_coef)*(1-zero_inf)*away_prb[1])]
        lst1 = away_prb[2::]
        for i in lst1:
            away_prb1.append(i*(1-draw_coef)*(1-zero_inf))
        

        prod_table = np.array([(i*j) for i, j in product(home_prb1, away_prb1)])
        prod_table.shape = (10, 10)

        prob_df = pd.DataFrame(prod_table, index=range(10), columns=range(10))
        return prob_df
    def prb_chances(home_team,away_team):
        data = new_prb.prb_matrix(home_team,away_team)
        home_win = sum((data.iat[1,0]+data.iat[2,0]+data.iat[3,0]+data.iat[4,0]+data.iat[5,0]+data.iat[6,0]+data.iat[7,0]+data.iat[8,0]+data.iat[9,0],
        data.iat[2,1]+data.iat[3,1]+data.iat[4,1]+data.iat[5,1]+data.iat[6,1]+data.iat[7,1]+data.iat[8,1]+data.iat[9,1],
        data.iat[3,2]+data.iat[4,2]+data.iat[5,2]+data.iat[6,2]+data.iat[7,2]+data.iat[8,2]+data.iat[9,2],
        data.iat[4,3]+data.iat[5,3]+data.iat[6,3]+data.iat[7,3]+data.iat[8,3]+data.iat[9,3],
        data.iat[5,4]+data.iat[6,4]+data.iat[7,4]+data.iat[8,4]+data.iat[9,4],
        data.iat[6,5]+data.iat[7,5]+data.iat[8,5]+data.iat[9,5],
        data.iat[7,6]+data.iat[8,6]+data.iat[9,6],
        data.iat[8,7]+data.iat[9,7]+data.iat[9,8]))*100
        draw = (data.iat[0,0]+data.iat[1,1]+data.iat[2,2]+data.iat[3,3]+data.iat[4,4]+data.iat[5,5]+data.iat[6,6]+data.iat[7,7]+data.iat[8,8]+data.iat[9,9])*100
        away_win = sum((data.iat[0,1]+data.iat[0,2]+data.iat[0,3]+data.iat[0,4]+data.iat[0,5]+data.iat[0,6]+data.iat[0,7]+data.iat[0,8]+data.iat[0,9],
        data.iat[1,2]+data.iat[1,3]+data.iat[1,4]+data.iat[1,5]+data.iat[1,6]+data.iat[1,7]+data.iat[1,8]+data.iat[1,9],
        data.iat[2,3]+data.iat[2,4]+data.iat[2,5]+data.iat[2,6]+data.iat[2,7]+data.iat[2,8]+data.iat[2,9],
        data.iat[3,4]+data.iat[3,5]+data.iat[3,6]+data.iat[3,7]+data.iat[3,8]+data.iat[3,9],
        data.iat[4,5]+data.iat[4,6]+data.iat[4,7]+data.iat[4,8]+data.iat[4,9],
        data.iat[5,6]+data.iat[5,7]+data.iat[5,8]+data.iat[5,9],
        data.iat[6,7]+data.iat[6,8]+data.iat[6,9],
        data.iat[7,8]+data.iat[7,9]+data.iat[8,8]))*100
        return float("{:.2f}".format(home_win)), float("{:.2f}".format(draw)), float("{:.2f}".format(away_win))
```
